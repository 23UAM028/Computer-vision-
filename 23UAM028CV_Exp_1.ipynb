```python
import cv2 as cv
import sys
from google.colab.patches import cv2_imshow
```

```python
image_path = '/content/drive/MyDrive/CV/img.jpg'
image = cv.imread(image_path)
cv2_imshow(image)

alpha = 1 # Contrast control (1.0-3.0)
beta = 1 # Brightness control (0-100)
adjusted = cv.convertScaleAbs(image, alpha=alpha, beta=beta)
# cv2_imshow(adjusted)

blur1 = cv.GaussianBlur(image, (5,5), 3)
cv2_imshow(blur1)

blur2 = cv.GaussianBlur(image, (5,5),5,1)
cv2_imshow(blur2)

blur3 = cv.GaussianBlur(image, (5,5),0)
cv2_imshow(blur3)

blur4 = cv.GaussianBlur(image, (5,5),1,1)
cv2_imshow(blur4)

half = cv.resize(image, (0, 0), fx = 0.1, fy = 0.1)
cv2_imshow(half)
```
